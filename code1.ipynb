```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import csv

# Set the random seed
np.random.seed(2001)

# Generate time series data
t = np.arange(1, 201)
y_white_noise = np.random.normal(loc=0, scale=1, size=200)

# Plot the simulated time series, ACF, and PACF
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot the simulated time series
axes[0].plot(t, y_white_noise, label="White Noise")
axes[0].set_xlabel("Time")
axes[0].set_ylabel("y_t")
axes[0].set_title("Simulated White Noise")

# Plot the sample ACF
plot_acf(y_white_noise, lags=20, ax=axes[1])
axes[1].set_title("Sample ACF")

# Plot the sample PACF
plot_pacf(y_white_noise, lags=20, ax=axes[2], method='ywm')
axes[2].set_title("Sample PACF")

plt.tight_layout()
plt.show()

# Create a time series with different frequencies
yt_monthly = pd.Series(y_white_noise, index=pd.date_range(start="1960-01", periods=200, freq="M"))
yt_yearly = pd.Series(y_white_noise, index=pd.date_range(start="1960", periods=200, freq="Y"))
yt_quarterly = pd.Series(y_white_noise, index=pd.date_range(start="1960-01", periods=200, freq="Q"))

# Plot the time series with different frequencies
fig, axes = plt.subplots(3, 1, figsize=(10, 12))

axes[0].plot(yt_monthly)
axes[0].set_title("Monthly Frequency")

axes[1].plot(yt_yearly)
axes[1].set_title("Yearly Frequency")

axes[2].plot(yt_quarterly)
axes[2].set_title("Quarterly Frequency")

plt.tight_layout()
plt.show()

# Load CO2 dataset
co2 = pd.read_csv("co2.csv", index_col=0, parse_dates=True)
co2_series = co2.dropna()

# Plot the CO2 dataset
plt.figure(figsize=(10, 5))
plt.plot(co2_series, label="Original CO2")
plt.title("CO2 Levels")
plt.show()

# Compute and plot ACF and PACF
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

plot_acf(co2_series, lags=100, ax=axes[0])
axes[0].set_title("Sample ACF")

plot_pacf(co2_series, lags=100, ax=axes[1], method='ywm')
axes[1].set_title("Sample PACF")

plt.tight_layout()
plt.show()

# Smoothing by moving average
co2_ma = co2_series.rolling(window=12, center=True).mean()
plt.plot(co2_series, label="Original CO2")
plt.plot(co2_ma, label="Smoothed Series", color='red')
plt.legend()
plt.title("CO2 Smoothed Series")
plt.show()

# First and second-order differencing
co2_ma_1stdiff = co2_ma.diff().dropna()
co2_ma_2stdiff = co2_ma_1stdiff.diff().dropna()

# Plot the differences
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

axes[0].plot(co2_ma_1stdiff, label="First Difference", color='green')
axes[0].set_title("First Difference of Smoothed Series")

axes[1].plot(co2_ma_2stdiff, label="Second Difference", color='purple')
axes[1].set_title("Second Difference of Smoothed Series")

plt.tight_layout()
plt.show()

# Plot ACF and PACF for the differenced series
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

plot_acf(co2_ma_1stdiff.dropna(), lags=100, ax=axes[0])
axes[0].set_title("Sample ACF of First Differenced Series")

plot_pacf(co2_ma_1stdiff.dropna(), lags=100, ax=axes[1], method='ywm')
axes[1].set_title("Sample PACF of First Differenced Series")

plt.tight_layout()
plt.show()

# Combined plots for original, differenced, and smoothed
fig, axes = plt.subplots(3, 1, figsize=(10, 12))

axes[0].plot(co2_series, label="Original CO2")
axes[0].set_title("Original CO2 Series")

axes[1].plot(co2_ma_1stdiff, label="First Difference", color='green')
axes[1].set_title("First Difference")

axes[2].plot(co2_ma, label="Smoothed Series", color='red')
axes[2].set_title("Smoothed Series")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Helper function to add ACF and PACF plots
def add_acf_pacf_subplot(data, lags, ax_acf, ax_pacf, title_prefix=""):
    plot_acf(data, lags=lags, ax=ax_acf)
    ax_acf.set_title(f"{title_prefix} ACF")
    plot_pacf(data, lags=lags, ax=ax_pacf, method='ywm')
    ax_pacf.set_title(f"{title_prefix} PACF")

# Set the random seed
np.random.seed(2001)

# Generate white noise time series
t = np.arange(1, 201)
y_white_noise = np.random.normal(loc=0, scale=1, size=200)

# time series with different frequencies
yt_monthly = pd.Series(y_white_noise, index=pd.date_range(start="1960-01", periods=200, freq="M"))
yt_yearly = pd.Series(y_white_noise, index=pd.date_range(start="1960", periods=200, freq="Y"))
yt_quarterly = pd.Series(y_white_noise, index=pd.date_range(start="1960-01", periods=200, freq="Q"))

# Load and process CO2 dataset (replace with your actual path)( by the way this is a R dataset)
co2 = pd.read_csv("co2.csv", index_col=0, parse_dates=True).dropna()
co2_ma = co2.rolling(window=12, center=True).mean()
co2_ma_1stdiff = co2_ma.diff().dropna()
co2_ma_2stdiff = co2_ma_1stdiff.diff().dropna()

# Create figure dynamically
fig, axes = plt.subplots(8, 2, figsize=(18, 24))  # Adjust grid as needed
axes = axes.ravel()  # Flatten to iterate easily

# Plot each series
axes[0].plot(t, y_white_noise, label="White Noise")
axes[0].set_title("Simulated White Noise")
axes[0].set_xlabel("Time")
axes[0].set_ylabel("y_t")

add_acf_pacf_subplot(y_white_noise, lags=20, ax_acf=axes[1], ax_pacf=axes[2], title_prefix="White Noise")

axes[3].plot(yt_monthly)
axes[3].set_title("Monthly Frequency")

axes[4].plot(yt_yearly)
axes[4].set_title("Yearly Frequency")

axes[5].plot(yt_quarterly)
axes[5].set_title("Quarterly Frequency")

axes[6].plot(co2)
axes[6].set_title("Original CO2 Levels")

add_acf_pacf_subplot(co2, lags=100, ax_acf=axes[7], ax_pacf=axes[8], title_prefix="CO2")

axes[9].plot(co2, label="Original CO2")
axes[9].plot(co2_ma, label="Smoothed", color='red')
axes[9].legend()
axes[9].set_title("Smoothed CO2 Levels")

axes[10].plot(co2_ma_1stdiff, color='green')
axes[10].set_title("First Difference")

axes[11].plot(co2_ma_2stdiff, color='purple')
axes[11].set_title("Second Difference")

add_acf_pacf_subplot(co2_ma_1stdiff, lags=100, ax_acf=axes[12], ax_pacf=axes[13], title_prefix="First Difference")

axes[14].plot(co2, label="Original CO2")
axes[14].set_title("Original CO2 Series")

axes[15].plot(co2_ma_1stdiff, label="First Difference", color='green')
axes[15].set_title("First Difference Combined")

plt.tight_layout()
plt.show()
```
